<b>b) Observe the unique number of values, most repeated values, and least repeated values of the variables.</b>

```python
# Function to calculate unique, most repeated and least repeated values for each column
def unique_most_least_repeated(df):
    unique_counts = df.nunique()
    most_repeated = df.mode().iloc[0]
    least_repeated = df.apply(lambda x: x.value_counts().index[-1])

    summary = pd.DataFrame({
        'Unique # of Values': unique_counts,
        'Most Repeated Values': most_repeated,
        'Least Repeated Values': least_repeated

    })

    return summary

# Applying the function to the dataframe
unique_most_least_summary = unique_most_least_repeated(df_Hi_renamed)

unique_most_least_summary
```

<b>Observations</b>

Unique Values: This column shows the total number of unique values that each variable takes. For example, 'At-Bats' has 247 unique values, and 'League' has only 2, which makes sense since 'League' is a categorical variable with two categories.

Most Repeated Value: This column indicates the value that occurs most frequently in each variable. For instance, 'Home-Runs' most often has a value of 4, which could be the mode for that variable.

Least Repeated Value: This column shows the value that occurs least frequently for each variable. These values could be considered outliers in some cases. For example, the least repeated value for 'Career Hits' is 1457, which could be a high outlier indicating a player with an exceptionally successful career.

<b> 2)  How do we impute the missing values for df_Hi? Why did you use that, write your reasoning.  </b>

<b> K-Nearest Neighbors (KNN) Imputation: </b>

This method imputes missing values based on the nearest neighbors found in the multidimensional space of other features.
It is more sophisticated and can yield better results if there's a pattern or correlation between 'Salary' and other variables.

<b> 3) Verify that you correctly imputed the missing values.  </b>

```python
from sklearn.impute import KNNImputer

# Creating the KNN imputer with default parameters (n_neighbors=5)
imputer = KNNImputer(n_neighbors=5)

# Applying the imputer to the dataframe. We should use only numerical columns for KNN imputation.
df_numerical = df_Hi_renamed.select_dtypes(include=['int64', 'float64'])
df_imputed = imputer.fit_transform(df_numerical)

# Creating a new dataframe with the imputed values
df_Hi_renamed_imputed = df_Hi_renamed.copy()
df_Hi_renamed_imputed[df_numerical.columns] = df_imputed

# Checking if the missing values in the Salary column have been imputed
missing_values_after_imputation = df_Hi_renamed_imputed['Salary'].isna().sum()
missing_values_after_imputation
```

```python
# Import the display functionality
from IPython.display import display, HTML

# Create a DataFrame with the summary statistics before imputation
salary_stats_before_df = df_Hi_renamed['Salary'].describe().to_frame(name='Before Imputation')

# Create a DataFrame with the summary statistics after imputation
salary_stats_after_df = df_Hi_renamed_imputed['Salary'].describe().to_frame(name='After Imputation')

# Combine both DataFrames for a side-by-side comparison
comparison_df = salary_stats_before_df.join(salary_stats_after_df)

# Renaming the index to match the 'Salary' column title
comparison_df.index.name = 'Salary'

# Use IPython.display to show the DataFrame
display(comparison_df)
```

    A correlation of 1 indicates a perfect positive relationship between two variables, where they move in the same direction.

In the heatmap, darker red colors indicate higher positive correlations, and darker blue colors indicate higher negative correlations. Lighter colors or white indicate no correlation or a very weak correlation.

<b> d) Observe the correlation coefficients and identify the pairs that has a correlation of more than 0.8 in absolute value.  </b>

```python
# Identifying pairs with a correlation coefficient of more than 0.8 in absolute value

high_correlation_pairs = correlation_matrix.abs().stack().reset_index()

high_correlation_pairs.columns = ['Variable 1', 'Variable 2', 'Correlation']

high_correlation_pairs = high_correlation_pairs[high_correlation_pairs['Variable 1'] != high_correlation_pairs['Variable 2']]  # Remove self-correlation

high_correlation_pairs = high_correlation_pairs[high_correlation_pairs['Correlation'] > 0.8].sort_values(by='Correlation', ascending=False)



# Since the matrix is symmetric, we'll drop duplicate pairs

high_correlation_pairs = high_correlation_pairs.drop_duplicates(subset=['Correlation'])

high_correlation_pairs.reset_index(drop=True, inplace=True)



high_correlation_pairs
```

<b>Observations:</b>

'Division_W' and 'Division_E' have a perfect negative correlation of -1.0, which is expected since each team can only be in one division, making these variables complementary.

'New_League_N' and 'New_League_A' have a perfect negative correlation of -1.0 for the same reason as above.

Career statistics like 'Career At Bats', 'Career Hits', 'Career Home-Runs', 'CareerRuns', and 'Career Runs Batted In' have very high correlations with each other, suggesting that players who are good in one aspect tend to be good in others, and these statistics accumulate together over a player's career.

'At-Bats' and 'Hits', 'Hits' and 'Runs', as well as other pairs related to hitting performance, also show high correlations, indicating that players who get to bat more often tend to have more hits and consequently score more runs.
